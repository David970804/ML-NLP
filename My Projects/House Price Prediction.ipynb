### Part 2 Pipeline building with data preprocessing and regression estimator


```python
# target variable: median_house_value
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
# base class for user-defined transformer
from sklearn.base import BaseEstimator, TransformerMixin
# seperately transform each group of columns and concatenate them in the end
from sklearn.compose import ColumnTransformer

# add a few variables of interest
# housing["rooms_per_household"] = housing["total_rooms"]/housing["households"]
# housing["bedrooms_per_room"] = housing["total_bedrooms"]/housing["total_rooms"]
# housing["population_per_household"] = housing["population"]/housing["households"]
```

```python
# pipeline order: (simpleimputer ->add columns -> standardize numerical values) + onehotencoder for categorical values
# using columntransformer to do the trick

# missing values imputer
imp = SimpleImputer(missing_values=np.nan, strategy='median')

# attributes combiner
# column index
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6
class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True): # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self  # nothing else to do
    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                         bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

# pipeline for numerical variables
num_pipeline = Pipeline([
    ('imputer', imp),
    ('var_combiner',CombinedAttributesAdder()),
    ('std_scaler',StandardScaler())
])
```

```python
housing_features = housing.drop(columns=['income_cat','median_house_value'])
housing_target = housing[['median_house_value']]
housing_features.info()
```

```python
# Build data preprocessor

col_names = list(housing_features) # return list of column names
cat_attribs = ["ocean_proximity"]
num_attribs = [var for var in col_names if var != 'ocean_proximity']

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(drop='first'), cat_attribs)
])
```

```python
# how to get names of categorical transformers
full_pipeline.fit(housing_features).transformers_[1][1]\
.get_feature_names()
#OneHotEncoder().fit_transform(housing_reg[['ocean_proximity']]).toarray()
len(full_pipeline.transform(housing_features)[0]) # 15 features including numerical and categorical
```

```python
# combine preprocessors and estimator
from sklearn.linear_model import Lasso

lasso_reg = Pipeline(steps=[
    ('preprocessor',full_pipeline),
    ('estimator', Lasso())
])

len(lasso_reg.fit(housing_features,housing_target).steps[1][1].coef_) # 1 intercept + 15 coefficients
```

### Part3 Hyperparameter tuning with Grid Search

```python
from sklearn.model_selection import GridSearchCV

alpha_space = np.logspace(-3,2,100)
param = {'estimator__alpha':alpha_space}
lasso_cv = GridSearchCV(estimator=lasso_reg, param_grid=param, cv=5)
```

```python
# train test split and fit lasso_cv
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(housing_features,housing_target,
                                                   test_size = 0.25, random_state = 42, stratify = None)
lasso_cv.fit(X_train, y_train)
```

```python
print('best parameter: ',lasso_cv.best_params_)
print('best score: ', lasso_cv.best_score_)
```

```python
# using tuned alpha on the test set
from sklearn.metrics import mean_squared_error
lasso_test = Pipeline([
    ('processor', full_pipeline),
    ('estimator', Lasso(alpha=100))
])

lasso_test.fit(X_test, y_test)
y_pred = lasso_test.predict(X_test)
print('fitting score:',lasso_test.score(X_test, y_test))
print('mean square error',mean_squared_error(y_test, y_pred)) 
```
